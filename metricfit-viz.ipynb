```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
npz = np.load('save/fm_fold0_cost1.npz')
entr, metric, reward = npz['entr_tr'], npz['metric_tr'], npz['wcost_tr']-npz['scost_tr']
```

```python
eidx = np.argsort(entr)
plt.Figure()
plt.plot(entr[eidx[::10]], reward[eidx[::10]], 'oC4', markersize=4, alpha=0.25, label='Reward')
plt.plot(entr[eidx], metric[eidx], '-C3', label="Metric")
plt.grid(); plt.legend(); plt.xlabel('Entropy h(x)'); plt.ylabel('Metric m(x) / Reward R(x,y)')
plt.show()
```

```python
msort = np.sort(metric)
plt.figure()
plt.plot(msort, np.linspace(0, 1, msort.shape[0]))
plt.grid(); plt.xlabel('Metric value m'); plt.ylabel('Ratio of images with m(x) < m')
plt.show()
```
